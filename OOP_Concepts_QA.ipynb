## Q1. What are the Four Pillars of OOP? 🟢

**Definition:** The four pillars are Encapsulation, Inheritance, Polymorphism, and Abstraction.

**Key Takeaway:** Encapsulation protects data, Inheritance reuses code, Polymorphism provides flexibility, Abstraction hides complexity.

**Benefits of OOP:**
- **Modularity**: Objects are standalone entities communicating through public interfaces, promoting code separation and easier maintenance
- **Reusability**: Class hierarchies and objects can be reused across projects; inheritance allows acquiring attributes from parent classes
- **Extensibility**: New features can be added through inheritance and interfaces without modifying existing code
- **Flexibility**: Polymorphism enables objects to adapt behavior based on context with the same interface

```python
from abc import ABC, abstractmethod

# 1. Encapsulation - data hiding
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private
    
    def get_balance(self):
        return self.__balance

# 2. Inheritance - code reuse
class SavingsAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.interest_rate = interest_rate

# 3. Polymorphism - same interface, different behavior
class Dog:
    def sound(self): return "Bark"

class Cat:
    def sound(self): return "Meow"

# 4. Abstraction - hiding complexity
class Shape(ABC):
    @abstractmethod
    def area(self): pass

# Test
account = BankAccount(1000)
print(f"Balance: {account.get_balance()}")

savings = SavingsAccount(2000, 0.05)
print(f"Savings Balance: {savings.get_balance()}")

def make_sound(animal):
    print(animal.sound())

make_sound(Dog())
make_sound(Cat())
```

## Q2. Classes and Objects 🟢

**Definition:** A class is a blueprint; an object is an instance of a class.

**Key Takeaway:** Class defines structure, object is the actual entity.

```python
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no
    
    def display(self):
        print(f"Student: {self.name}, Roll: {self.roll_no}")

# Creating objects
student1 = Student("Amit", 101)
student2 = Student("Priya", 102)
student1.display()
student2.display()
```

## Q3. Abstraction in OOP 🟢

**Definition:** Hiding complex implementation and showing only essential features.

**Key Takeaway:** Abstraction defines "what to do", concrete classes define "how to do it".

```python
from abc import ABC, abstractmethod

class Database(ABC):
    @abstractmethod
    def connect(self):
        pass
    
    @abstractmethod
    def execute_query(self, query):
        pass

class MySQL(Database):
    def connect(self):
        print("Connected to MySQL")
    
    def execute_query(self, query):
        print(f"Executing MySQL query: {query}")

class MongoDB(Database):
    def connect(self):
        print("Connected to MongoDB")
    
    def execute_query(self, query):
        print(f"Executing MongoDB query: {query}")

# Test
db = MySQL()
db.connect()
db.execute_query("SELECT * FROM users")
```

## Q4. Encapsulation 🟢

**Definition:** Bundling data and methods, restricting direct access to components.

**Key Takeaway:** Use `__` for private members, provide public methods for controlled access.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    
    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(f"Balance: {account.get_balance()}")
# print(account.__balance)  # Would raise AttributeError
```

## Q5. Inheritance Types 🟢

**Types:** Single, Multiple, Multilevel, Hierarchical, Hybrid

**Key Takeaway:** Child class inherits all accessible members from parent.

```python
# Single Inheritance
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    
    def start(self):
        print(f"{self.brand} vehicle started")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    
    def drive(self):
        print(f"Driving {self.brand} {self.model}")

# Multiple Inheritance
class Father:
    def skills(self): print("Programming")

class Mother:
    def talent(self): print("Dancing")

class Child(Father, Mother):
    pass

# Test
my_car = Car("Toyota", "Camry")
my_car.start()
my_car.drive()

child = Child()
child.skills()
child.talent()
```

## Q8. Types of Polymorphism 🟡

**Types:** Runtime (Method Overriding) and Compile-time (Method Overloading)

**Key Takeaway:** Python primarily supports runtime polymorphism and duck typing.

```python
# Runtime Polymorphism (Method Overriding)
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

# Compile-time Polymorphism (via default args)
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

# Duck typing
def make_sound(animal):
    animal.sound()

# Test
make_sound(Dog())
make_sound(Cat())

calc = Calculator()
print(f"add(5): {calc.add(5)}")
print(f"add(5, 3): {calc.add(5, 3)}")
print(f"add(5, 3, 2): {calc.add(5, 3, 2)}")
```

## Q9. Magic Methods (Dunder Methods) 🟡

**Definition:** Special methods with double underscores that enable operator overloading.

**Key Takeaway:** Magic methods let custom classes work with Python's built-in operations.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
    
    def __str__(self):
        return f"Book: {self.title}"
    
    def __len__(self):
        return self.pages
    
    def __add__(self, other):
        return self.pages + other.pages

book1 = Book("Python Basics", 300)
book2 = Book("Advanced Python", 400)
print(book1)
print(f"Length: {len(book1)}")
print(f"Total pages: {book1 + book2}")
```

## Q10. __init__ vs __new__ 🟡

**Definition:** `__new__` creates the instance, `__init__` initializes it.

**Key Takeaway:** `__new__` runs first and must return an instance; `__init__` runs after.

```python
class Person:
    def __new__(cls, name):
        print(f"Creating instance with __new__")
        instance = super().__new__(cls)
        return instance
    
    def __init__(self, name):
        print(f"Initializing instance with __init__")
        self.name = name

# Singleton pattern using __new__
class Singleton:
    _instance = None
    
    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

# Test
p = Person("Alice")

s1 = Singleton()
s2 = Singleton()
print(f"s1 is s2: {s1 is s2}")
```

## Q11. Class vs Instance Variables 🟡

**Definition:** Class variables are shared; instance variables are unique to each object.

**Key Takeaway:** Modifying class variable affects all instances.

```python
class Employee:
    company = "TechCorp"  # Class variable
    
    def __init__(self, name, salary):
        self.name = name      # Instance variable
        self.salary = salary

emp1 = Employee("Alice", 50000)
emp2 = Employee("Bob", 60000)

print(f"emp1 company: {emp1.company}")
print(f"emp2 company: {emp2.company}")

Employee.company = "NewTech"
print(f"After change - emp1: {emp1.company}")
print(f"After change - emp2: {emp2.company}")

print(f"emp1 name: {emp1.name}")
print(f"emp2 name: {emp2.name}")
```

## Q12. @staticmethod, @classmethod, instance methods 🟡

**Definition:** Instance methods use `self`, class methods use `cls`, static methods use neither.

**Key Takeaway:** Choose based on what data the method needs to access.

```python
class MathOperations:
    pi = 3.14159
    
    def __init__(self, number):
        self.number = number
    
    def square(self):  # Instance method
        return self.number ** 2
    
    @classmethod
    def get_pi(cls):  # Class method
        return cls.pi
    
    @staticmethod
    def add(x, y):  # Static method
        return x + y

obj = MathOperations(5)
print(f"Square: {obj.square()}")
print(f"Pi: {MathOperations.get_pi()}")
print(f"Add: {MathOperations.add(3, 4)}")
```

## Q13. Method Overriding 🟡

**Definition:** Child class provides specific implementation of parent's method.

**Key Takeaway:** Same method signature, different implementation.

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):  # Overriding
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):  # Overriding
        return 3.14 * self.radius ** 2

rect = Rectangle(5, 3)
circle = Circle(7)
print(f"Rectangle area: {rect.area()}")
print(f"Circle area: {circle.area()}")
```

## Q14. super() Function 🟡

**Definition:** Returns proxy object to call parent class methods.

**Key Takeaway:** Essential for proper initialization and multiple inheritance.

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print(f"Vehicle init: {brand}")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  # Call parent constructor
        self.model = model
        print(f"Car init: {model}")

# Multiple inheritance
class Electric:
    def charge(self):
        print("Charging battery")

class Hybrid(Car, Electric):
    def __init__(self, brand, model, battery):
        super().__init__(brand, model)
        self.battery = battery

car = Hybrid("Toyota", "Prius", "50kWh")
car.charge()
```

## Q21. Method Resolution Order (MRO) 🔴

**Definition:** Order in which Python searches for methods in class hierarchy.

**Key Takeaway:** Python uses C3 linearization for MRO.

```python
class A:
    def show(self):
        print("A's method")

class B(A):
    def show(self):
        print("B's method")

class C(A):
    def show(self):
        print("C's method")

class D(B, C):  # Multiple inheritance
    pass

obj = D()
obj.show()
print(f"MRO: {D.mro()}")
```

## Q23. @property Decorator 🔴

**Definition:** Allows methods to be accessed like attributes with validation.

**Key Takeaway:** Provides encapsulation with clean syntax.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

temp = Temperature(25)
print(f"Celsius: {temp.celsius}")
print(f"Fahrenheit: {temp.fahrenheit}")
temp.celsius = 30
print(f"New Celsius: {temp.celsius}")
```

## Q26. Singleton Pattern 🔴

**Definition:** Design pattern ensuring only one instance exists.

**Key Takeaway:** Use `__new__` or decorator to implement singleton.

```python
# Method 1: Using __new__
class Singleton:
    _instance = None
    
    def __new__(cls, *args, **kwargs):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
    
    def __init__(self, value):
        self.value = value

s1 = Singleton(10)
s2 = Singleton(20)
print(f"s1 is s2: {s1 is s2}")
print(f"s1.value: {s1.value}")
print(f"s2.value: {s2.value}")

# Method 2: Using decorator
def singleton(cls):
    instances = {}
    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return get_instance

@singleton
class DatabaseConnection:
    def __init__(self, host):
        self.host = host

db1 = DatabaseConnection("localhost")
db2 = DatabaseConnection("remote")
print(f"db1 is db2: {db1 is db2}")
```

## Q27. Shallow Copy vs Deep Copy 🔴

**Definition:** Shallow copy copies references; deep copy copies values recursively.

**Key Takeaway:** Use `deepcopy()` for nested objects.

```python
import copy

class Address:
    def __init__(self, city):
        self.city = city

class Person:
    def __init__(self, name, address):
        self.name = name
        self.address = address

# Original
addr = Address("Mumbai")
p1 = Person("Amit", addr)

# Shallow copy
p2 = copy.copy(p1)
p2.name = "Priya"
p2.address.city = "Delhi"  # Changes both!

print(f"p1 name: {p1.name}")
print(f"p1 city: {p1.address.city}")
print(f"p2 city: {p2.address.city}")

# Deep copy
p3 = copy.deepcopy(p1)
p3.address.city = "Bangalore"

print(f"After deep copy:")
print(f"p1 city: {p1.address.city}")
print(f"p3 city: {p3.address.city}")
```
