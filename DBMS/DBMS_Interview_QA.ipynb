## Q1. What is DBMS? 🟢

**Definition:** DBMS (Database Management System) is software that manages and organizes data in a structured way.

**Key Advantages:**
- Data Independence
- Data Sharing
- Data Security
- Reduced Redundancy

```python
# SQLite example for demonstration
import sqlite3

# Create database and table
conn = sqlite3.connect(':memory:')  # In-memory database
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE Books (
    book_id INTEGER PRIMARY KEY,
    title TEXT,
    author TEXT,
    published_year INTEGER
)
''')

# Insert data
cursor.execute("INSERT INTO Books VALUES (1, 'Python Basics', 'Alice', 2020)")
cursor.execute("INSERT INTO Books VALUES (2, 'Advanced SQL', 'Bob', 2021)")
conn.commit()

# Query data
cursor.execute("SELECT * FROM Books")
print("Books:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## Q2. Database vs DBMS vs RDBMS 🟢

| Aspect | Database | DBMS | RDBMS |
|--------|----------|------|-------|
| **What is it?** | Data collection | Management software | Relational management software |
| **Structure** | Any format | Files/tables | Tables with relationships |
| **Examples** | Excel file | MongoDB | MySQL, PostgreSQL |
| **ACID** | No | Partial | Full ACID compliance |

```python
# RDBMS example showing relationships
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create related tables
cursor.execute('''
CREATE TABLE Customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT
)
''')

cursor.execute('''
CREATE TABLE Orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    order_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id)
)
''')

# Insert data
cursor.execute("INSERT INTO Customers VALUES (1, 'Alice'), (2, 'Bob')")
cursor.execute("INSERT INTO Orders VALUES (101, 1, '2024-01-15'), (102, 2, '2024-01-16')")
conn.commit()

# Join query
cursor.execute('''
SELECT c.name, o.order_id, o.order_date
FROM Customers c
JOIN Orders o ON c.customer_id = o.customer_id
''')

print("Customer Orders:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## Q4. Types of SQL Commands 🟢

| Category | Commands | Purpose |
|----------|----------|---------||
| **DDL** | CREATE, ALTER, DROP, TRUNCATE | Define schema |
| **DML** | INSERT, UPDATE, DELETE | Manipulate data |
| **DCL** | GRANT, REVOKE | Control access |
| **TCL** | COMMIT, ROLLBACK | Manage transactions |
| **DQL** | SELECT | Query data |

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# DDL: Define structure
cursor.execute('CREATE TABLE Employees (id INTEGER, name TEXT, salary REAL)')
print("Table created (DDL)")

# DML: Manipulate data
cursor.execute("INSERT INTO Employees VALUES (1, 'John', 50000)")
cursor.execute("INSERT INTO Employees VALUES (2, 'Jane', 60000)")
print("Data inserted (DML)")

# DQL: Query data
cursor.execute("SELECT * FROM Employees WHERE salary > 45000")
print("\nQuery Results (DQL):")
for row in cursor.fetchall():
    print(row)

# DML: Update
cursor.execute("UPDATE Employees SET salary = 55000 WHERE id = 1")
print("\nData updated (DML)")

cursor.execute("SELECT * FROM Employees")
print("After update:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## Q5. Primary Key vs Foreign Key 🟢

| Aspect | Primary Key | Foreign Key |
|--------|-------------|-------------|
| **Purpose** | Uniquely identifies records | Creates relationships |
| **Uniqueness** | Must be unique | Can have duplicates |
| **NULL** | Cannot be NULL | Can be NULL |
| **Count** | One per table | Multiple allowed |

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Primary Key
cursor.execute('''
CREATE TABLE Departments (
    dept_id INTEGER PRIMARY KEY,
    dept_name TEXT
)
''')

# Foreign Key
cursor.execute('''
CREATE TABLE Employees (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT,
    dept_id INTEGER,
    FOREIGN KEY (dept_id) REFERENCES Departments(dept_id)
)
''')

# Insert data
cursor.execute("INSERT INTO Departments VALUES (1, 'IT'), (2, 'HR')")
cursor.execute("INSERT INTO Employees VALUES (101, 'Alice', 1), (102, 'Bob', 2)")
conn.commit()

# Query with join
cursor.execute('''
SELECT e.emp_name, d.dept_name
FROM Employees e
JOIN Departments d ON e.dept_id = d.dept_id
''')

print("Employee-Department Relationship:")
for row in cursor.fetchall():
    print(f"{row[0]} works in {row[1]}")

conn.close()
```

## Q9. ACID Properties 🟢

**ACID ensures reliable transaction processing:**
- **A**tomicity: All or nothing
- **C**onsistency: Valid state transitions
- **I**solation: Concurrent transactions don't interfere
- **D**urability: Committed data persists

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create accounts table
cursor.execute('''
CREATE TABLE Accounts (
    account_id TEXT PRIMARY KEY,
    balance REAL
)
''')

cursor.execute("INSERT INTO Accounts VALUES ('A', 1000), ('B', 500)")
conn.commit()

print("Initial balances:")
cursor.execute("SELECT * FROM Accounts")
for row in cursor.fetchall():
    print(row)

# Transaction: Transfer $100 from A to B
try:
    cursor.execute("BEGIN TRANSACTION")
    
    # Atomicity: Both operations must succeed
    cursor.execute("UPDATE Accounts SET balance = balance - 100 WHERE account_id = 'A'")
    cursor.execute("UPDATE Accounts SET balance = balance + 100 WHERE account_id = 'B'")
    
    conn.commit()  # Durability: Changes persist
    print("\nTransaction committed successfully!")
except Exception as e:
    conn.rollback()  # Atomicity: Rollback if any fails
    print(f"Transaction failed: {e}")

print("\nFinal balances:")
cursor.execute("SELECT * FROM Accounts")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## Q11. SELECT, WHERE, ORDER BY 🟢

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE Employees (
    id INTEGER,
    name TEXT,
    department TEXT,
    salary REAL,
    age INTEGER
)
''')

employees = [
    (1, 'Alice', 'IT', 75000, 28),
    (2, 'Bob', 'HR', 55000, 32),
    (3, 'Charlie', 'IT', 85000, 25),
    (4, 'Diana', 'Finance', 70000, 30)
]

cursor.executemany('INSERT INTO Employees VALUES (?,?,?,?,?)', employees)
conn.commit()

# SELECT with WHERE
print("IT Department employees:")
cursor.execute("SELECT name, salary FROM Employees WHERE department = 'IT'")
for row in cursor.fetchall():
    print(row)

# ORDER BY
print("\nEmployees ordered by salary (DESC):")
cursor.execute("SELECT name, salary FROM Employees ORDER BY salary DESC")
for row in cursor.fetchall():
    print(row)

# Combined
print("\nIT employees earning > 60000, ordered by salary:")
cursor.execute('''
SELECT name, salary 
FROM Employees 
WHERE department = 'IT' AND salary > 60000
ORDER BY salary DESC
''')
for row in cursor.fetchall():
    print(row)

conn.close()
```

## Q12. Aggregate Functions 🟢

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('CREATE TABLE Employees (id INTEGER, name TEXT, salary REAL, department TEXT)')
employees = [
    (1, 'Alice', 75000, 'IT'),
    (2, 'Bob', 55000, 'HR'),
    (3, 'Charlie', 85000, 'IT'),
    (4, 'Diana', 70000, 'Finance')
]
cursor.executemany('INSERT INTO Employees VALUES (?,?,?,?)', employees)

# COUNT
cursor.execute("SELECT COUNT(*) as total FROM Employees")
print(f"Total employees: {cursor.fetchone()[0]}")

# AVG, MIN, MAX, SUM
cursor.execute('''
SELECT 
    AVG(salary) as avg_salary,
    MIN(salary) as min_salary,
    MAX(salary) as max_salary,
    SUM(salary) as total_payroll
FROM Employees
''')

result = cursor.fetchone()
print(f"\nAverage Salary: ${result[0]:.2f}")
print(f"Minimum Salary: ${result[1]:.2f}")
print(f"Maximum Salary: ${result[2]:.2f}")
print(f"Total Payroll: ${result[3]:.2f}")

conn.close()
```

## Q13. GROUP BY vs HAVING 🟢

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('CREATE TABLE Employees (name TEXT, department TEXT, salary REAL)')
employees = [
    ('Alice', 'IT', 75000),
    ('Bob', 'HR', 55000),
    ('Charlie', 'IT', 85000),
    ('Diana', 'Finance', 70000),
    ('Eve', 'IT', 65000),
    ('Frank', 'HR', 60000)
]
cursor.executemany('INSERT INTO Employees VALUES (?,?,?)', employees)

# GROUP BY
print("Average salary by department:")
cursor.execute('''
SELECT department, COUNT(*) as emp_count, AVG(salary) as avg_salary
FROM Employees
GROUP BY department
''')
for row in cursor.fetchall():
    print(f"{row[0]}: {row[1]} employees, avg ${row[2]:.2f}")

# HAVING - filter groups
print("\nDepartments with more than 2 employees:")
cursor.execute('''
SELECT department, COUNT(*) as emp_count
FROM Employees
GROUP BY department
HAVING COUNT(*) > 2
''')
for row in cursor.fetchall():
    print(f"{row[0]}: {row[1]} employees")

conn.close()
```

## Q16. Types of Joins 🟡

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create tables
cursor.execute('CREATE TABLE Employees (emp_id INTEGER, name TEXT, dept_id INTEGER)')
cursor.execute('CREATE TABLE Departments (dept_id INTEGER, dept_name TEXT)')

cursor.execute("INSERT INTO Employees VALUES (1, 'Alice', 10), (2, 'Bob', 20), (3, 'Charlie', NULL)")
cursor.execute("INSERT INTO Departments VALUES (10, 'IT'), (20, 'HR'), (30, 'Finance')")

# INNER JOIN
print("INNER JOIN (matching only):")
cursor.execute('''
SELECT e.name, d.dept_name
FROM Employees e
INNER JOIN Departments d ON e.dept_id = d.dept_id
''')
for row in cursor.fetchall():
    print(f"  {row[0]} - {row[1]}")

# LEFT JOIN
print("\nLEFT JOIN (all employees):")
cursor.execute('''
SELECT e.name, d.dept_name
FROM Employees e
LEFT JOIN Departments d ON e.dept_id = d.dept_id
''')
for row in cursor.fetchall():
    dept = row[1] if row[1] else 'None'
    print(f"  {row[0]} - {dept}")

conn.close()
```

## Q19. Subqueries 🟡

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('CREATE TABLE Employees (name TEXT, salary REAL, dept TEXT)')
employees = [
    ('Alice', 75000, 'IT'),
    ('Bob', 55000, 'HR'),
    ('Charlie', 85000, 'IT'),
    ('Diana', 45000, 'HR')
]
cursor.executemany('INSERT INTO Employees VALUES (?,?,?)', employees)

# Subquery: Employees earning above average
print("Employees earning above average:")
cursor.execute('''
SELECT name, salary
FROM Employees
WHERE salary > (SELECT AVG(salary) FROM Employees)
''')
for row in cursor.fetchall():
    print(f"  {row[0]}: ${row[1]}")

# Correlated subquery
print("\nEmployees earning above their department average:")
cursor.execute('''
SELECT e1.name, e1.dept, e1.salary
FROM Employees e1
WHERE e1.salary > (
    SELECT AVG(e2.salary)
    FROM Employees e2
    WHERE e2.dept = e1.dept
)
''')
for row in cursor.fetchall():
    print(f"  {row[0]} ({row[1]}): ${row[2]}")

conn.close()
```
