## Q7. Mutable vs Immutable 🟢

**Mutable:** Can be changed (list, dict, set)
**Immutable:** Cannot be changed (int, str, tuple)

```python
# Mutable - List
my_list = [1, 2, 3]
my_list[0] = 10
print(f"Modified list: {my_list}")

# Immutable - String
text = "hello"
# text[0] = "H"  # TypeError
new_text = text.upper()  # Creates new string
print(f"Original: {text}, New: {new_text}")

# Important: Mutable default argument pitfall
def add_item_bad(item, lst=[]):
    lst.append(item)
    return lst

print(f"Call 1: {add_item_bad(1)}")
print(f"Call 2: {add_item_bad(2)}")  # Unexpected!

# Correct way
def add_item_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(f"Good Call 1: {add_item_good(1)}")
print(f"Good Call 2: {add_item_good(2)}")
```

## Q9. *args and **kwargs 🟢

**Definition:** `*args` passes variable positional arguments, `**kwargs` passes variable keyword arguments.

```python
# *args - variable positional arguments
def sum_all(*args):
    return sum(args)

print(f"Sum: {sum_all(1, 2, 3, 4, 5)}")

# **kwargs - variable keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="NYC")

# Combined
def flexible_function(a, b, *args, **kwargs):
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"kwargs={kwargs}")

flexible_function(1, 2, 3, 4, 5, x=10, y=20)
```

## Q21. Decorators 🟡

**Definition:** Functions that modify behavior of other functions.

```python
import time
from functools import wraps

# Basic decorator
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()

# Timing decorator
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.1)
    return "Done"

result = slow_function()

# Decorator with arguments
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(times=3)
def greet(name):
    print(f"Hello, {name}!")

greet("Python")
```

## Q22. Generators and Yield 🟡

**Definition:** Functions that return an iterator using `yield`.

```python
# Basic generator
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)

# Generator expression
squares_gen = (x**2 for x in range(10))
print(f"First 3 squares: {[next(squares_gen) for _ in range(3)]}")

# Fibonacci generator
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib_list = list(fibonacci(10))
print(f"Fibonacci: {fib_list}")

# Memory efficient file reading
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

## Q25. Shallow vs Deep Copy 🟡

**Shallow copy:** Copies object, shares references to nested objects
**Deep copy:** Recursively copies all objects

```python
import copy

# Original list with nested list
original = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow = copy.copy(original)
shallow[0][0] = 999  # Modifies original too!

print(f"Original after shallow copy: {original}")
print(f"Shallow copy: {shallow}")

# Deep copy
original = [[1, 2, 3], [4, 5, 6]]
deep = copy.deepcopy(original)
deep[0][0] = 999  # Does NOT modify original

print(f"\nOriginal after deep copy: {original}")
print(f"Deep copy: {deep}")

# Assignment (just reference)
original = [1, 2, 3]
reference = original
reference[0] = 999
print(f"\nOriginal with reference: {original}")
```

## Q28. Multithreading vs Multiprocessing 🟡

**Multithreading:** Multiple threads in one process (I/O-bound tasks)
**Multiprocessing:** Multiple processes (CPU-bound tasks)

```python
import threading
import multiprocessing
import time

# Multithreading example
def task(name):
    print(f"Thread {name} starting")
    time.sleep(0.1)
    print(f"Thread {name} finished")

threads = []
for i in range(3):
    t = threading.Thread(target=task, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("All threads completed")

# Multiprocessing example
def cpu_task(n):
    return sum(i*i for i in range(n))

if __name__ == '__main__':
    with multiprocessing.Pool(processes=2) as pool:
        results = pool.map(cpu_task, [1000000, 2000000])
        print(f"Results: {results}")
```

## Q29. GIL (Global Interpreter Lock) 🟡

**Definition:** Mutex that allows only one thread to execute Python bytecode at a time.

**Key Takeaway:** GIL limits true parallelism in CPython for CPU-bound tasks.

```python
import threading
import time

# CPU-bound task (affected by GIL)
def cpu_bound_task():
    count = 0
    for i in range(10000000):
        count += i
    return count

# Single-threaded
start = time.time()
cpu_bound_task()
cpu_bound_task()
single_time = time.time() - start
print(f"Single-threaded: {single_time:.2f}s")

# Multi-threaded (not much faster due to GIL)
start = time.time()
t1 = threading.Thread(target=cpu_bound_task)
t2 = threading.Thread(target=cpu_bound_task)
t1.start()
t2.start()
t1.join()
t2.join()
multi_time = time.time() - start
print(f"Multi-threaded: {multi_time:.2f}s")
print("Note: Multi-threading may not be faster due to GIL!")
```

## Q35. Regular Expressions 🟡

**Definition:** Pattern matching for strings using the `re` module.

```python
import re

# Basic matching
text = "My phone is 123-456-7890"
pattern = r'\d{3}-\d{3}-\d{4}'
match = re.search(pattern, text)
print(f"Found: {match.group() if match else 'Not found'}")

# Find all matches
text = "Emails: test@example.com, user@domain.org"
emails = re.findall(r'[\w.-]+@[\w.-]+', text)
print(f"Emails: {emails}")

# Substitution
text = "Hello World"
new_text = re.sub(r'World', 'Python', text)
print(f"Substituted: {new_text}")

# Groups
text = "Date: 2024-01-15"
match = re.search(r'(\d{4})-(\d{2})-(\d{2})', text)
if match:
    year, month, day = match.groups()
    print(f"Year: {year}, Month: {month}, Day: {day}")

# Compile for reuse
phone_pattern = re.compile(r'\d{3}-\d{3}-\d{4}')
print(f"Is valid phone: {bool(phone_pattern.match('123-456-7890'))}")
```

## Q36. zip() and enumerate() 🟡

## Q8. Pass by Reference or Value? 🟢

**Definition:** Python uses pass by object reference.

**Key Takeaway:** Modifying mutable objects affects original; reassignment creates new object.

```python
# Immutable - appears like pass by value
def modify_immutable(x):
    x = 100
    print(f"Inside: {x}")

num = 10
modify_immutable(num)
print(f"Outside: {num}")  # Still 10

# Mutable - appears like pass by reference
def modify_mutable(lst):
    lst.append(4)
    print(f"Inside: {lst}")

my_list = [1, 2, 3]
modify_mutable(my_list)
print(f"Outside: {my_list}")  # Modified!

# Reassignment doesn't affect original
def reassign_list(lst):
    lst = [99, 99]
    print(f"Inside after reassign: {lst}")

my_list = [1, 2, 3]
reassign_list(my_list)
print(f"Outside after reassign: {my_list}")  # Unchanged
```

## Q17. Global vs Local Variables 🟢

**Local:** Inside functions
**Global:** Outside functions

**Key Takeaway:** Use local by default, `global` keyword when necessary.

```python
# Global and local scope
global_var = "I'm global"

def my_function():
    local_var = "I'm local"
    print(f"Access global: {global_var}")
    print(f"Local: {local_var}")

my_function()

# Modifying global
count = 0

def increment():
    global count
    count += 1

increment()
increment()
print(f"\nCount: {count}")

# LEGB Rule
x = "global"

def outer():
    x = "enclosing"
    
    def inner():
        x = "local"
        print(f"Inner: {x}")
    
    inner()
    print(f"Outer: {x}")

outer()
print(f"Global: {x}")
```

## Q23. Iterators and Iterables 🟡

**Iterable:** Has `__iter__()`
**Iterator:** Has `__iter__()` and `__next__()`

**Key Takeaway:** All iterators are iterables.

```python
# Iterable vs Iterator
my_list = [1, 2, 3]
my_iter = iter(my_list)

print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3

# Custom Iterator
class Counter:
    def __init__(self, start, end):
        self.current = start
        self.end = end
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        self.current += 1
        return self.current - 1

print("\nCustom Counter:")
for num in Counter(1, 5):
    print(num, end=' ')
```

**Definition:** Manage resources with `with` statement.

**Key Takeaway:** Ensures cleanup even when exceptions occur.

```python
# Custom context manager
class DatabaseConnection:
    def __init__(self, db_name):
        self.db_name = db_name
    
    def __enter__(self):
        print(f"Connecting to {self.db_name}")
        return self
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Closing {self.db_name}")
        return False

with DatabaseConnection("mydb") as db:
    print("Using database")

# Using contextlib
from contextlib import contextmanager
import time

@contextmanager
def timer(name):
    start = time.time()
    yield
    print(f"{name} took {time.time() - start:.4f}s")

with timer("Operation"):
    time.sleep(0.1)
    print("Working...")
```
